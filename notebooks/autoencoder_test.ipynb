```python
import torch
import torch.nn as nn
import sys
from pathlib import Path

# Project root path setup
current_path = Path.cwd()

if current_path.name == "notebooks":
    project_root = current_path.parent
else:
    project_root = current_path

sys.path.append(str(project_root))

from models.autoencoder import FeatureAutoencoder
```

```python
# Shared project settings
batch_size = 32
sequence_length = 10
cnn_feature_dim = 128
latent_dim = 64

# Dummy CNN feature sequence
dummy_cnn_features = torch.randn(batch_size, sequence_length, cnn_feature_dim)

print("Dummy CNN feature shape:", dummy_cnn_features.shape)
```

```python
autoencoder = FeatureAutoencoder(
    input_dim=cnn_feature_dim,
    latent_dim=latent_dim,
    dropout_rate=0.3
)

latent, reconstructed = autoencoder(dummy_cnn_features)

print("Input shape:        ", dummy_cnn_features.shape)
print("Latent shape:       ", latent.shape)
print("Reconstructed shape:", reconstructed.shape)
```

```python
criterion = nn.MSELoss()
reconstruction_loss = criterion(reconstructed, dummy_cnn_features)

print("Reconstruction loss:", reconstruction_loss.item())
```

```python
assert dummy_cnn_features.shape == torch.Size([32, 10, 128])
assert latent.shape == torch.Size([32, 10, 64])
assert reconstructed.shape == torch.Size([32, 10, 128])

print("Autoencoder shape test passed successfully.")
```
